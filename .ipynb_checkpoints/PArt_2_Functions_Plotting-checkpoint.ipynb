```python
#Part2 Intro to Python
#Data visualization using Seaborn
import pandas as pd
import seaborn as sns
sns.set(color_codes = True)
```

```python
#import Auotmotive dataset
ship = pd.read_csv('titanic.csv')
```

```python
ship.head()
```

```python
sns.distplot(ship['Survived']) #remeber only int type are taken in the distplot i.e histogram
```

```python
sns.distplot(ship['Pclass'], kde=False,rug=True)
```

```python
#scatterplot
    sns.jointplot(ship['Age'],ship['Fare'], kind = "hex")
```

```python
#pairplot 
sns.pairplot(ship[['Age', 'Fare', 'SibSp']])
```

```python
#stripplot
sns.stripplot(ship['Sex'], ship['Age'])
```

```python
sns.stripplot(ship['Sex'], ship['Age'], jitter = True)
```

```python
#swarmplot
sns.swarmplot(ship['Sex'], ship['Age'])
```

```python
#boxplot
sns.boxplot(ship['Sex'], ship['Age'], hue = ship['Survived'])
```

```python
#barplot
sns.barplot(ship['Sex'], ship['Age'], hue = ship['Survived'])
```

```python
#countplot
sns.countplot(ship['Embarked'])
```

```python
#pointplot
sns.pointplot(ship['Sex'], ship['Age'], hue = ship['Pclass'])
```

```python
sns.factorplot(x="Sex", 
               y="Age", 
               hue="Fare", 
               col="Survived", 
               data=ship, 
               kind="swarm")
```

```python
#regression
sns.lmplot(x="Survived", y="Fare", data=ship)
```

```python
#functions
def comparo(num1, num2):
    if(num1 > num2):
        print("Greater number is ",num1)
        return num1
    if(num2 > num1):
        print("Greater number is ",num2)
        return num2
    else:
        print("numbers equal")
    
comparo(8,10)
```

```python
sum = lambda x,y: (x + y)  # The function has no name. Lambda returns a function object referenced
                           # as double in this example

# Output: 10
print(sum(5,6))
```
